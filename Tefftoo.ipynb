```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.plotly as py  
import plotly.tools as tls   
from plotly.graph_objs import *
```

```python
f, f_err, r, r_err, T, T_err = np.genfromtxt("data/log.dat").T
fig, ax = plt.subplots(figsize=(12,8))
ax.errorbar(f, r, xerr=f_err, yerr=r_err, fmt="k.", capsize=0, alpha=.5, ecolor=".6")
plt.ylabel("density")
plt.xlabel("flicker")
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.errorbar(T, r, xerr=T_err, yerr=r_err, fmt="k.", capsize=0, alpha=.5, ecolor=".6")
plt.xlabel("Teff")
plt.ylabel("density")
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.errorbar(T, f, xerr=T_err, yerr=f_err, fmt="k.", capsize=0, alpha=.5, ecolor=".6")
plt.xlabel("Teff")
plt.ylabel("flicker")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(f, r, T)
ax.set_xlabel("flicker")
ax.set_ylabel("density")
ax.set_zlabel("Teff")
ax.view_init(20, 60)
```
